Check the `student_scores.csv` file you just create, and print out the answers to the following 3 questions:
- who gets the highest score after calculation, and what is the score?
- who gets the lowest score after calculation, and what is the score?
- what is the average score among all students? Round the results to 2 decimal places.

```python
### INSERT YOUR CODE HERE
import os
def analyze_scores(filename):
    max_score = -float('inf')
    min_score = float('inf')
    total = 0.0
    count = 0
    top_students = []
    bottom_students = []

    if not os.path.exists(filename):
      print(f"Error: file {filename}do not exist.")
      return None

    with open(filename, 'r') as f:
      for line in f:
          line = line.strip()
          if not line:
              continue

          if ',' not in line:
              print(f"Invalid format skipped: {line}")
              continue

          parts = line.split(',', 1)
          if len(parts) != 2:
              print(f"Invalid format skipped: {line}")
              continue

          name, score_str = parts

          if not score_str.replace('.', '', 1).isdigit():
              print(f"Invalid score skipped: {name} - {score_str}")
              continue

          score = float(score_str)

          count += 1
          total += score

          if score > max_score:
              max_score = score
              top_students = [name]
          elif score == max_score:
              top_students.append(name)

          if score < min_score:
              min_score = score
              bottom_students = [name]
          elif score == min_score:
              bottom_students.append(name)

    if count == 0:
        print("Warning: No valid data found")
        return None

    average = round(total / count, 2)

    return {
        "top_students": (top_students, max_score),
        "lowest_students": (bottom_students, min_score),
        "average_score": average
    }

#for output testing
results = analyze_scores("student_scores.csv")

if results:
  print(f"[Highest Score] {', '.join(results['top_students'][0])} : {results['top_students'][1]:.2f}")
  print(f"[Lowest Score] {', '.join(results['lowest_students'][0])} : {results['lowest_students'][1]:.2f}")
  print(f"[Average Score] {results['average_score']:.2f}")
else:
  print("No valid results generated")

```

### Task 1
Write a function `is_outlier()` that can check for certain kinds of outliers. It should check that:
1. if the demonstrated interest score is 0 or
2. if the normalized GPA that is more than 2 points higher than the normalized SAT score.

If either of these conditions is true, it should return `True` (because this student is an outlier); otherwise, the function returns `False`.

```python
### INSERT YOUR CODE HERE
def is_outlier(featrue_list):
  sat=featrue_list[0]
  gpa=featrue_list[1]
  interest=featrue_list[2]

  return (interest==0) or ((sat/160) - (gpa*2) > 2)

```

### Task 1
Write a function `grade_outlier()` that takes in a list of grades (of any length) and returns `True` if one single number is more than 20 points lower than all other numbers; otherwise, `False`.

Example:

> Input: [99, 94, 87, 89, 56, 78, 89]

Hint: Sort the list from lowest to highest, and check for the difference between the two lowest grades.

78 - 56 = 22; 22 > 20

> Output: True

Next, consider the data that we have: a list of grades for each student, one grade per semester for four semesters.

Make sure that your `grade_outlier()` function works by calling it for every row in the second dataset. Print out an informative message about which students have a single grade outlier. You'll delete this later but it's a great way of testing your function!

Finally, consider the importance of an algorithm being able to flag students who might have a lower overall GPA but have shown improvement over time.

```python
### INSERT YOUR CODE HERE
def grade_outlier(grades):
  if len(grades) < 2:
      return False

  sorted_grades = sorted(grades)
  if (sorted_grades[1] - sorted_grades[0]) > 20:
    return True
  return False
```

### Task 2
Create a function `grade_improvement()` that returns `True` if the average score of each semester is higher than or equal to each previous semester and `False` otherwise.

Hint: investigate how the `==` operator works between two lists and think about using the Python's `sorted()` function.

```python
### INSERT YOUR CODE HERE
def grade_improvement(grades):
  sorted_grades = sorted(grades)
  if sorted_grades==grades:
    return True
  return False
```
